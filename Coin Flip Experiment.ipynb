## COIN FLIP EXPERIMENT

This project was created for learning purposes to explore the relationship between experimental and true probabilities of flipping a specified amount of heads in various scenarios.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import pandas as pd
```

This function calculates probability of flipping the exact number of heads out of a pre-set amount of coin flips, using a randomly generated sample. The experiment results are then compared with the probability of the same event, calculated using a binomial distribution function.

```python
def coin_flip_experiment(num_of_flips, num_of_trials, H):
    coins = []
    proportion = []
    trials = []
    successful_outcomes = 0
    trial = 1
    for flip in range(num_of_trials):
        coin_set = np.random.choice(['H', 'T'], num_of_flips, p=[0.5, 0.5])
        coins.append(list(coin_set))
    for subset in coins:
        if subset.count('H') == H:
            successful_outcomes += 1
            proportion.append(successful_outcomes / trial)
            trials.append(trial)
        else: 
            proportion.append(successful_outcomes / trial)
            trials.append(trial)
        trial += 1
    experimental_probability = successful_outcomes / num_of_trials    
    true_probability = binom.pmf(k=H, n=num_of_flips, p=0.5)
    
    plt.plot(trials, proportion, label='Experimental Probability')
    plt.title(f'Flipping {H} Heads Out Of {num_of_flips} Flips')
    plt.xlabel('Number of Trials')
    plt.ylabel('Proportion of Successful Outcomes')
    plt.hlines(true_probability, 0, num_of_trials, colors='orange', label='True Probability')
    plt.legend()
    plt.show()
      
    return print('Experimental probability: ', round(experimental_probability,4), '\nTrue probability: ', 
                 round(true_probability,4))
```

```python
coin_flip_experiment(6, 5000, H=2)
```

This function calculates and visualizes a probability distribution of flipping from 0 to specified amount of heads out of a pre-set amount of coin flips, using a randomly generated sample. The experiment results are then compared with the true probability distribution of the same events, calculated using a binomial distribution function.

```python
def coin_flip_experiment_2_1(num_of_flips, num_of_trials, H):
    coins = []
    proportion = []
    trials = []
    successful_outcomes = 0
    trial = 1
    for flip in range(num_of_trials):
        coin_set = np.random.choice(['H', 'T'], num_of_flips, p=[0.5, 0.5])
        coins.append(list(coin_set))
    for subset in coins:
        if subset.count('H') == H:
            successful_outcomes += 1
            proportion.append(successful_outcomes / trial)
            trials.append(trial)
        else: 
            proportion.append(successful_outcomes / trial)
            trials.append(trial)
        trial += 1
    experimental_probability = successful_outcomes / num_of_trials    

    return experimental_probability


def coin_flip_experiment_2(num_of_flips, num_of_trials):
    probability = []
    trial = list(i for i in range(num_of_flips + 1))
    for i in range(num_of_flips + 1):
        result = coin_flip_experiment_2_1(num_of_flips, num_of_trials, i)
        probability.append(result)

    plt.bar(x=trial, height=probability, width=0.8)
    plt.title(f'Experimental probability distribution \nof flipping from {0} to {num_of_flips} heads in {num_of_flips} flips \nin {num_of_trials} trials')
    plt.ylabel('Proportion of Successful Outcomes')
    plt.xlabel(f'Number of Heads In {num_of_flips} Flips')
    plt.show()
    plt.clf()

    probability = []
    for i in range(num_of_flips + 1):
        result = binom.pmf(k=i, n=num_of_flips, p=0.5)
        probability.append(result)

    plt.bar(x=trial, height=probability, width=0.8)
    plt.title(f'True probability distribution \nof flipping from {0} to {num_of_flips} heads in {num_of_flips} flips')
    plt.ylabel('Probability')
    plt.xlabel(f'Number of Heads In {num_of_flips} Flips')
    plt.show()
    plt.clf()
    
```

This function calculates probablity of flipping at least a given number of heads out of pre-set amount of coin flips, using randomly generated sample. The experiment results are then compared with the probability of the same event calculated using binomial distribution function. 

```python
def coin_flip_experiment_3(num_of_flips, num_of_trials, H):
    coins = []
    proportion = []
    trials = []
    successful_outcomes = 0
    trial = 1
    for flip in range(num_of_trials):
        coin_set = np.random.choice(['H', 'T'], num_of_flips, p=[0.5, 0.5])
        coins.append(list(coin_set))
    for subset in coins:
        if subset.count('H') >= H:
            successful_outcomes += 1
            proportion.append(successful_outcomes / trial)
            trials.append(trial)
        else: 
            proportion.append(successful_outcomes / trial)
            trials.append(trial)
        trial += 1
    experimental_probability = successful_outcomes / num_of_trials    
    true_probability = 1 - binom.cdf(k=H-1, n=num_of_flips, p=0.5)
        
    plt.plot(trials, proportion, label='Experimental Probability')
    plt.title(f'Flipping at least {H} Heads Out Of {num_of_flips} Coins')
    plt.xlabel('Number of Trials')
    plt.ylabel('Proportion of Successful Outcome')
    plt.hlines(true_probability, 0, num_of_trials, colors='orange', label='True Probability')
    plt.legend()
    plt.show()

    return print('Experimental probability: ', round(experimental_probability,4), '\nTrue probability: ', round(true_probability,4))
```

This function calculates probablity of flipping at least a given number of heads in a row out of pre-set amount of coin flips, using randomly generated sample.

```python
def n_in_a_row(n, List, letter):
    if n == 0:
        for i in List:
            if i == letter:
                return False
            else: return True
    for a in range(len(List)):
        Count = []
        if List[a] == letter:
            Count.append(1)
            for b in range((a+1),len(List)):
                if List[a] == List[b]:
                    Count.append(1)  
                    if sum(Count) == n:
                        return True
                    else:
                        continue
                elif sum(Count) == n:
                    return True
                else:
                    Count.clear()
                    continue
        else: 
            continue
    return False      

def coin_flip_experiment_4(num_of_flips, num_of_trials, H):
    coins = []
    proportion = []
    trials = []
    successful_outcomes = 0
    trial = 1
    for flip in range(num_of_trials):
        coin_set = np.random.choice(['H', 'T'], num_of_flips, p=[0.5, 0.5])
        coins.append(list(coin_set))
    for subset in coins:
        if subset.count('H') >= H:
            if n_in_a_row(H, subset, 'H') == True:
                successful_outcomes += 1
                proportion.append(successful_outcomes / trial)
                trials.append(trial)
            else:
                proportion.append(successful_outcomes / trial)
                trials.append(trial)
        else: 
            proportion.append(successful_outcomes / trial)
            trials.append(trial)
        trial += 1
    experimental_probability = successful_outcomes / num_of_trials
    
    return print('Experimental probability of flipping ', H, 'or more heads in a row: ', round(experimental_probability,4))
```
